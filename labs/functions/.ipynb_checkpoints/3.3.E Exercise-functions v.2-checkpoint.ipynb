#### Q1. 
Write a function that takes two numbers, `a` and `b`, converts them to `int`, and then returns the result of multiplying `a` by the negative of `b`.

```python
# Remember: a and b are called parameters!

# your code here
def multibly(a,b):
    a = int(a)
    b = int (b)
    return a*b
```

#### Q3. 
Write a function that takes a collection, converts it to a `tuple` with all duplicates removed and returns the result.

```python
# your code here
def q3_answer(list1):
    list1 = set(list1)
    list1 = tuple(list1)
    return list1
```

```python
q3_answer(['a', 'b', 'a', 'a', 'b', 'c'])
```

##### Q5.C
Write a function that accepts one parameter, a list
The function should return the sum total of all numbers in the list, using a loop.
\
*Hint*: Do we want a for loop or a while loop for this kind of question?

```python
# your code here
def sum_list(listt):
    summ = 0
    for i in listt:
       summ += i
    return summ
```

```python
my_list=[3,7,1,9]

sum_list(my_list)
```

#### Q6. Problem-Solving Functions

```python
# A. Write a function that finds the longest string in a given list of strings

# your code here
def longest_string_finder(dictionary):
    lenth = ''
    for i in dictionary:
        if len(i) > len(lenth):
            lenth = i
    return lenth
```

```python
s=["dfaf","adfaffaf","grrtsrfsgfgfsgsgf","adfaaa"]

longest_string_finder(s)
```

```python
# B. Write a function to determine whether a given list contains a given value

# your code here
def search_list(listt, value):
    if value in listt:
        return True
    else:
        return False 
```

```python
ls=[34,90,120,432,647,875]

search_list(ls,432)
```

```python
ls=[34,90,120,432,647,875]

search_list(ls,122)
```

```python
# C. Write a function to find the area of a triangle given its length and height as parameters

# your code here
def area(lenght, height):
    a = 0.5*lenght*height
    return a
```

```python
area(4,6)
```

```python
# D. Write a function to count how many times a given string contains a given letter

# your code here
def char_count(text, letter):
    count = 0
    for i in text:
       if i == letter:
           count+= 1  
    return count
```

```python
s2="This is a test about data science short course with excellent students"

char_count(s2,"i")
```

#### Q7. Challange Question (Optional)  

```python
# A. Write a function to find the most common letter in a given string, and use the previous function in your answer
#    Note: do not include the space " " as a character

# your code here
def most_common_letter(text):
    count = 0
    mmax = ' '
    for i in text:
       if char_count(text, i) > count and i != ' ':
           count = char_count(text, i)
           mmax = i
    return mmax
```

```python
s3="This is a test about data science short course with excellent students"

most_common_letter(s3)
```
