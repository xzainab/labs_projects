#### 6.3 GroupBy Insights
Segment the above dataframe and explore aggregate values to answer the following questions

```python
# A. Which discount results in the highest mean ordered quantity?
orders.groupby('discount')['quantity'].mean().sort_values(ascending=False)
```

```python
# 6.4 Joining DataFrames
# The below example joins our first two dataframes by their shared column, Symbol
openprice = pd.DataFrame({'Symbol': ['AAPL', 'DHR', 'DAL', 'AMZN'], 'OpenPrice': [217.51, 96.54, 51.45, 1703.34]})
wkhigh = pd.DataFrame({'Symbol': ['DAL', 'AMZN', 'AAPL', 'DHR'], '52wkHigh': [60.79, 2050.49, 233.47, 110.11]})
combined = pd.merge(left = openprice, 
                    right = wkhigh, 
                    how="left", 
                    left_on='Symbol', 
                    right_on='Symbol')
combined.head()
```

```python
# A. Join the stockname dataframe to our combined result and print the result
stockname = pd.DataFrame({'Symbol': ['AMZN', 'DHR', 'DAL', 'AAPL'], 'Name': ['Amazon', 'Danaher', 'Delta Airlines', 'Apple']})
# complete the below code:
combined = pd.merge(left= combined, 
                    right= stockname, 
                    how="left", 
                    on="Symbol")
combined.head()
```

```python
# B. Use the following tables from Super Store
products = pd.read_csv('products.csv',encoding='latin')
orders = pd.read_csv('orders.csv')
returns = pd.read_csv('returns.csv')
regions = pd.read_csv('regions.csv')
```

```python
# i. We want to join the products and orders dataframes. 
#    Explore both dataframes to identify the common column between them

#    Use a left join to combine the tables and save the result table as
#    a dataframe named orders_with_products

# Your code here please:

orders_with_products = pd.merge(left=products, right=orders, how="left", on="product_id")
orders_with_products.head(2)
```

```python
# ii. Left join the orders_with_products and returns dataframes
# call the new dataframe orders_with_products_and_returns

orders_with_products_and_returns = pd.merge(left=orders_with_products, right=returns, how="left", on="order_id")
orders_with_products_and_returns.head(2)
```

```python
# iii. Finally, add the region data to our combined dataframe
# do inner join of the previous dataframe with `regions` dataframe 
# name the new dataframe orders_with_products_and_returns_and_regions

# Your code here please:
orders_with_products_and_returns_and_regions = pd.merge(left=orders_with_products_and_returns, right=regions, how="inner", on="region_id")
orders_with_products_and_returns_and_regions.head()
```

```python
# iv. Let's use this combined dataframe to determine the salesperson generating the most profit
orders_with_products_and_returns_and_regions.groupby('salesperson')['profit'].sum().sort_values(ascending=False)
```
