
Or in other words, the remainder of dividing 9 by 5 is 4. 

Hint2: Use `for` loop along with conditional statements `if-else`

### 6. Iterate through the animals. Print out the animal name and the number of vowels in the name.

````python
animals = ['duck', 'rat', 'boar', 'elefant', 'mammoth', 'gazelle']
vowels = 'aeiou'
````

````python
# your code here
for animal in animals:
    count = 0
    for letter in animal:
        if letter in vowels:
           count += 1
    print(f'{animal} {count}')
````

### 7. Iterate from 1 to 30 with the following instructions:

1. If a number is divisible by 3, print 'fizz'. 
2. If a number is divisible by 5, print 'buzz'. 
3. If a number is both divisible by 3 and 5 print 'fizzbuzz'.
4. Otherwise, print just the number.

Hint: the first condition to check should be the third condition. Why?

````python
# your code here
for i in range(1,31):
    if i%3 == 0 and i%5 == 0:
        print('fizzbuzz')
    elif i%3 == 0:
        print('fizz')
    elif i%5 == 0:
        print('buzz')
    else:
        print(i)
````

### 8. Use a while loop to sum the integers from 1 to 100 inclusively.

````python
# your code here
num = 1
summ = 0
while num <= 100:
    summ = summ + num
    num += 1

print(summ)
````

### 10. Write a while loop that removes in each traversal the last element from my_list and stops as soon as my_list is empty. Print each removal of an item and print at the end when the list is empty

Hint1: You can use `.pop()` function to remove the last item in a list.

Hint2: An empty list returns "False" and it returns "True" if there is an item inside the list.

````python
my_list=[1,2,3,4,5,6,7,8,9]

# your code here
while my_list:
    removed_item = my_list.pop()
    print(removed_item)
    
print (my_list, "now the list is empty")
````
