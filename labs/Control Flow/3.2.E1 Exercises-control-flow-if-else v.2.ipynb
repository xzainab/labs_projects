### 1. Write an if-else statement for multiple conditions.

Print out these recommendations based on the weather conditions:

1. Temperature over 20 ˚C and rain: bring an umbrella.
2. Temperature less than or equal to 20 ˚C and rain: bring and umbrella and a jacket.
3. Temperature over 20 ˚C and sunshine: wear a t-shirt.
4. Temperature less than or equal to 20 ˚C and sunshine: bring a jacket.

```python
# input() function is used to take input from the user (as string).
# eval() function is used here to convert the user input from string to a number

temperature = eval(input('What is the temperature? '))
weather = input('What is the weather? (rain or sunshine) ')
```

```python
# your code here
if temperature > 20 and weather == 'rain':
    print('bring an umbrella.')
elif temperature <= 20 and weather == 'rain':
    print('bring and umbrella and a jacket.')
elif temperature > 20 and weather == 'sunshine':
    print('wear a t-shirt.')
elif temperature <= 20 and weather == 'sunshine':
    print('bring a jacket.')
```

### 3. Complex Conditions

```python
# A. Write the code that determines if the triangle is:
#      equalateral - all three sides are equal in length
#      scalene - all three sides are unequal in length
#      isosceles - two sides are the same length
# use print("The triangle is ") based on the conditions you have tested 
# example, print("This triangle is equalateral") if it is equalateral...

side_a = 5
side_b = 7
side_c = 5
ttype = ''
# your code here
if side_a == side_b == side_c:
    ttype ='equalateral'
elif side_a != side_b != side_c:
    ttype = 'scalene'
elif side_a == side_b or side_a == side_c or side_b == side_c:
    ttype = 'isosceles'

print(f'This triangle is {ttype}')
```

```python
# B. Calculate what season it is based upon this chart:
#      Dec 21 - Mar 19: Winter
#      Mar 20 - Jun 20: Spring
#      Jun 21 - Sep 21: Summer
#      Sep 22 - Dec 20: Fall
# Print the result as follows:
#      <month> <day> is in <season> 
# Note: The months are Jan,Feb,Mar,Apr,May,Jun,Jul,Aug,Sep,Oct,Nov,Dec
# Tip: you can print using:  print(f"{month} {day} is in Winter")  as an example

month = "Sep"
day = 25

# your code here
if (month == "Dec" and day >= 21) or (month in ["Jan", "Feb"]) or (month == "Mar" and day <= 19):
    season = "Winter"
elif (month == "Mar" and day >= 20) or (month in ["Apr", "May"]) or (month == "Jun" and day <= 20):
    season = "Spring"
elif (month == "Jun" and day >= 21) or (month in ["Jul", "Aug"]) or (month == "Sep" and day <= 21):
    season = "Summer"
else:
    season = "Fall"

# 3. Print the result
print(f"{month} {day} is in {season}")
```

```python
# C. The Farmer's Boat
boat = {
    "has_fox": True,
    "has_chicken": False,
    "has_grain": True
}
# 1. If all three properties are true, print "The boat was too heavy and sank"
# 2. If the boat's has_fox and has_chicken properties are true, print "The fox ate the chicken" 
# 3. If the boat's has_chicken and has_grain propertes are true, print "The chicken ate the grain"
# 4. If none of them are true, print "The boat sank from disuse"
# 5. Otherwise, print "The boat is in a valid, productive state. Go boat!"

# your code here
if boat['has_fox'] == True and boat['has_chicken'] == True and boat['has_grain'] == True:
    print('The boat was too heavy and sank')
elif boat['has_fox'] == True and boat['has_chicken'] == True:
    print('The fox ate the chicken')
elif boat['has_chicken'] == True and boat['has_grain'] == True:
    print('The chicken ate the grain')
elif boat['has_fox'] == False and boat['has_chicken'] == False and boat['has_grain'] == False:
    print('The boat sank from disuse')
else:
    print('The boat is in a valid, productive state. Go boat!')

```
